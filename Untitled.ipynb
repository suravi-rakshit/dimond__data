```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_excel("diamonds.xlsx")
data=pd.DataFrame(df)
print(data.head())
grouped_by=df.groupby("cut")["depth"].sum()
print(grouped_by)
plt.bar(grouped_by.index,grouped_by.values)
plt.xlabel("cut",fontsize=20)
plt.ylabel("depth",fontsize=20)
plt.title("DIAMOND",fontsize=20)
plt.show()
```
